We need to generate feature maps of only convolution layers and not dense layers and hence we will generate feature maps of layers which have “dimension=4".

Each feature map has n-channels and this number “n” is given at the end of the shape of feature map. This is the number of features in a particular layer.
ES: shape:(1,64,64,8) is an image 64x64 with 8 dimensions, which are the 8 feature (images) combined together to form the final 8dimensional image. So, we need to iterate over this image to separate its 8 images. 

```python
def plot_feature_maps(k,layer_name,image_belt):
    scale = 20. / k
    plt.figure( figsize=(scale * k, scale) )
    plt.title ( layer_name )
    plt.grid  ( False )
    plt.imshow( image_belt, aspect='auto')
    

    for layer_name, feature_map in zip(layer_names, feature_maps):
    if len(feature_map.shape) == 4:
        k = feature_map.shape[-1]  
        size = feature_map.shape[1]
        for i in range(k):
            feature_image = feature_map[0, :, :, i]
            feature_image -= feature_image.mean()
            feature_image /= feature_image.std ()
            feature_image *=  64
            feature_image += 128
            feature_image = np.clip(x, 0, 255).astype('uint8')
            image_belt[:, i * size : (i+1)*size] = feature_image
            
            plot_feature_maps(k,layer_name,image_belt)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import random

def plot_losses(points,card):
    
    fig, ax = plt.subplots(figsize=(5, 5))

    points1,points2,points3 = points

    plt.scatter(card,points1)
    plt.scatter(card,points2)
    plt.scatter(card,points3)
    #plt.plot(card,points1,points2,points3)
    plt.plot(card, points1,  'b') # plotting t, a separately 
    plt.plot(card, points2, 'c') # plotting t, a separately 
    plt.plot(card, points3, 'r') # plotting t, a separately 

    ax.set_title('Losses')
    ax.set_xlabel('Epoch')
    ax.set_ylabel('values')
    plt.legend(["d_loss1","d_loss2","g_loss"])

    plt.xticks(np.arange(min(card), max(card)+1, 1))
    #plt.xticks(rotation=90)
    plt.grid()
    plt.show()
```
